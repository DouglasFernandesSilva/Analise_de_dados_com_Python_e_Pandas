```python
# Criando o dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
# Visualizando as 5 primerias linhas
df.head()
```

```python
# Quantidade de linhas e colunas
df.shape
```

```python
# Verificando os tipos de dados
df.dtypes
```

```python
# Qual a receita total?
df["Valor Venda"].sum()
```

```python
# Qual o custo total?
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"]) # Criando a coluna custo
```

```python
df.head(1)
```

```python
# QUal o custo total?
round(df["custo"].sum(), 2)
```

```python
# Agora que temos a receita e o custo total, podemos achar o lucro total
# Vamos criar uma coluna de lucro que será Receit - Custo
df["lucro"] = df["Valor Venda"] - df["custo"]
```

```python
df.head(1)
```

```python
# Total lucro
round(df["lucro"].sum(), 2)
```

```python
# Criando uma coluna com total de dias e para enviar o produto
df["tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

## **Agora, queremos saber a média do tempo de envio para cada Marca, e para isso precisamos transformar a coluna tempo_envio em numérica**

```python
# Extraindo apenas os dias
df["tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
# Verificando o tipo da coluna tempo_envio
df["tempo_envio"].dtype
```

```python
# Média do tempo de envia por marca
df.groupby("Marca")["tempo_envio"].mean()
```

# **Missing Values**

# **E se a gente quiser saber o lucro por ano e por marca?**

```python
# Vamos agrupar por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
```

```python
# Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
# Gráfico total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total pordutos vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
# Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title = "Lucro X Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```
